```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('data.csv')
```

```python
data['Date'] = pd.to_datetime(data['Date'])
```

### Будуємо графік функції і дивимось на перший погляд чи є сезонність і яка вона (адитивна чи мультиплікативна)

```python
ax = data.plot(x='Date', y='Open', figsize=(12,6))
years = ['2017-01-01', '2018-01-01', '2019-01-01', '2020-01-01','2021-01-01']
for i in years:
    plt.axvline(x=i, color='black', linestyle='--')
```

### Застосуємо модель seasonal_decompose щоб побачити частини нашого ряду (тренд, сезонність і залишок)
#### Так як я не знаю яка тут сезонність ( ад. чи мульт.) застосую 2 моделі

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
data.set_index('Date', inplace = True)
```

#### Multiplicative

```python
result_mul = seasonal_decompose(data['Open'], model='multiplicative', extrapolate_trend='freq')

#some data to research the correlation
trend = result_mul.trend
seasonal = result_mul.seasonal
residual = result_mul.resid

plt.rcParams.update({'figure.figsize': (20,20)})
result_mul.plot()
plt.show()
```

```python
result_add = seasonal_decompose(data['Open'], model='additive', extrapolate_trend='freq')

#some data to research the correlation
trend = result_add.trend
seasonal = result_add.seasonal
residual = result_add.resid

plt.rcParams.update({'figure.figsize': (20,20)})
result_add.plot()
plt.show()
```
